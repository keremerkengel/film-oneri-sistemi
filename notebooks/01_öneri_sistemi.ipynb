```python
import pandas as pd

# Yeni veri dosyalarını latin-1 ile oku
ratings = pd.read_csv("../data/ratings.dat", sep="::", engine="python",
                      names=["user_id", "movie_id", "rating", "timestamp"],
                      encoding="latin-1")

movies = pd.read_csv("../data/movies.dat", sep="::", engine="python",
                     names=["movie_id", "title", "genres"],
                     encoding="latin-1")

# Birleştir
data = pd.merge(ratings, movies, on="movie_id")

print("✅ Veri başarıyla yüklendi!")
data.head()
```

```python
# Kullanıcı-Film matrisini oluştur
user_movie_matrix = data.pivot_table(index="user_id", columns="title", values="rating")

print("✅ Kullanıcı-Film matrisi boyutu:", user_movie_matrix.shape)
user_movie_matrix.head()
```

```python
# Her sütun (film) arasındaki korelasyonu al
movie_similarity = user_movie_matrix.corr(method='pearson', min_periods=50)  # en az 50 ortak kullanıcıya bak

print("✅ Benzerlik matrisi oluşturuldu!")
movie_similarity.head()
```

```python
# Korelasyon matrisi (film-film arası)
movie_similarity = user_movie_matrix.corr(method='pearson', min_periods=30)

print("✅ Benzerlik matrisi oluşturuldu!")
movie_similarity.head()
```

```python
def film_öner(film_adi, kaç_tane=10):
    if film_adi not in movie_similarity:
        return f"⚠️ '{film_adi}' adlı film bulunamadı."
    
    benzer_filmler = movie_similarity[film_adi].dropna().sort_values(ascending=False)
    return benzer_filmler.iloc[1:kaç_tane+1]  # kendisini çıkar
```

```python
film_öner("Toy Story (1995)", 5)
```
